```python
# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Target Search Term
target_terms = ("@BBC", "@CBS", "@CNN", "@FOX", "@nytimes")
```

```python
# Array to hold sentiment
sentiment_array = []
sentiment_array_ALL = []
Media_Compounds_Mean = []

# Variable for holding the oldest tweet
oldest_tweet = ""
counter = 1

# Loop through all target users
for target in target_terms:
    print(target)
    
    # Run search around each tweet
    public_tweets = api.search(target, count=100, result_type="recent")
    #public_tweets = api.user_timeline(target)

    # Loop through all tweets
    for tweet in public_tweets["statuses"]:
       
        #print (json.dumps(tweet, indent = 4))
        # Run Vader Analysis on each tweet
        compound = analyzer.polarity_scores(tweet["text"])["compound"]
        pos = analyzer.polarity_scores(tweet["text"])["pos"]
        neu = analyzer.polarity_scores(tweet["text"])["neu"]
        neg = analyzer.polarity_scores(tweet["text"])["neg"]
            
        # Add sentiments for each tweet into an array
        sentiment_array.append({"Name": target, 
                                "Compound": compound,
                                "Positive": pos,
                                "Negative": neu,
                                "Neutral": neg,
                                "Tweet Text":tweet["text"],
                                "Tweets Ago": counter})
        
        sentiment_array_ALL.append({"Name": target, 
                                "Compound": compound,
                                "Positive": pos,
                                "Negative": neu,
                                "Neutral": neg,
                                "Tweet Text":tweet["text"],
                                "Tweets Ago": counter})
        counter = counter + 1
            
    sentiments_pd = pd.DataFrame.from_dict(sentiment_array)
    sentiments_pd.to_csv(target + "sentiments.csv")
        
    sentiment_array = []
    
    #Create plot for each News Agency
    plt.plot(np.arange(len(sentiments_pd["Compound"])),
                       sentiments_pd["Compound"], marker="o", linewidth=0.5,
                       alpha=0.8, label = "target")

    # Incorporate the other graph properties
    plt.title("Sentiment Analysis of Tweets (%s) for %s" % (time.strftime("%x"), target))
    plt.ylabel("Tweet Polarity")
    plt.xlabel("Tweets Ago")
    plt.show()          
    plt.savefig(target+".png")
    
    
sentiments_pd_ALL = pd.DataFrame.from_dict(sentiment_array_ALL)       
sentiments_pd_ALL.head(500)
```
